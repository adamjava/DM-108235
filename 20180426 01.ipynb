```python
from sklearn.datasets import load_iris
iris = load_iris()

X = iris.data
y = iris.target

from sklearn.model_selection import train_test_split
X_tr, X_te, y_tr, y_te = train_test_split(X, y, test_size = .5)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
clfknn = KNeighborsClassifier()
clftree = DecisionTreeClassifier()
```

```python
clfknn.fit(X_tr, y_tr) #uczenie się: dopasowywanie współczynników beta 
clftree.fit(X_tr, y_tr)

pred_knn = clfknn.predict(X_te)
pred_tree = clftree.predict(X_te)

print(pred_knn)
print(pred_tree)
print(y_te)

from sklearn.metrics import accuracy_score
print("knn: ", accuracy_score(y_te, pred_knn))
print("tree: ", accuracy_score(y_te, pred_tree))
```

```python
class TestDM():
    def fit(self, X_train, y_train):
        self.X_train = X_train
        self.y_train = y_train
    def predict(self, X_test):
        predictions = []
        for row in X_test:
            #label = 2 #zawsze zwracamy tą samą wartość
            label = np.random.choice(self.y_train) #klasyfikator zwraca losową wartość z labela
            predictions.append(label)
        return predictions
```

```python
clfDM  = TestDM()
clfDM .fit(X_tr, y_tr)
y_DM  = clfDM .predict(X_te)
print(accuracy_score(y_te, y_DM))
```

```python
from scipy.spatial import distance
def euc(a,b):
    return distance.euclidean(a,b)

#klasyfikator 1NN
class TestDM():
    def fit(self, X_train, y_train):
        self.X_train = X_train
        self.y_train = y_train
    def predict(self, X_test):
        predictions = []
        for row in X_test:
            label = self.closest(row)#znajdź najlbliższy punkt i weź jego klasę
            predictions.append(label)
        return predictions
    def closest(self,row):
        best_dist = euc(row,self.X_train[0])
        best_index = 0
        for i in range(1, len(self.X_train)):
            dist = euc(row, self.X_train[i])
            if dist < best_dist:
                best_dist = dist
                best_index = i
        return self.y_train[best_index]
    
clfDM  = TestDM()
clfDM .fit(X_tr, y_tr)
y_DM  = clfDM .predict(X_te)
print(accuracy_score(y_te, y_DM))
```
